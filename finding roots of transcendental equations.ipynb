```

import sympy as sp
def newton_raphson(x1,x2,func,error_tolerance=10^-2):
    root=(x1+x2)/2.0
    func_root=func.evalf(subs={x:root})
    derivative=sp.diff(func)
    der_value=derivative.evalf(subs={x:root})
    error=1000
    while abs(error)>error_tolerance:
        old_root=root
        root=root-(func_root/der_value)
        func_root=func.evalf(subs={x:root})
        der_value=derivative.evalf(subs={x:root})
        error=((root-old_root)/(root))*100
        print('old_root',old_root,'root',root,'error',error)

x=sp.Symbol('x')
func=-0.875*x*x+1.75*x+2.625
newton_raphson(-1.2,-0.9,func,10^-1)

    

    



```

```
import sympy as sp
def bisection_method(xinitial_1,xinitial_2,func,error_tolerance):
    x1=xinitial_1
    x2=xinitial_2
    x_r=(x1+x2)/2
    func_x1=func.evalf(subs={x:x1})
    func_x2=func.evalf(subs={x:x2})
    func_root=func.evalf(subs={x:x_r})
    error=1000
    while abs(error)>=error_tolerance:
        
        
        
        if (func_x1)*(func_root)<0:
            old_root=x_r
            x1=x1
         
           
            x2=x_r
            x_r=(x1+x2)/2
            func_x1=func.evalf(subs={x:x1})
            func_x2=func.evalf(subs={x:x2})
            func_root=func.evalf(subs={x:x_r})
            error=((x_r-old_root)/(x_r))*100
            print('x1=',x1   , 'x2=' , x2, 'x_r=',x_r  , abs(error),'\n')
        if abs(error)>=error_tolerance:
            if (func_x2)*(func_root)<0:
                old_root=x_r
                x2=x2
                x1=x_r
                x_r=(x1+x2)/2
                func_x1=func.evalf(subs={x:x1})
                func_x2=func.evalf(subs={x:x2})
                func_root=func.evalf(subs={x:x_r})
                error=((x_r-old_root)/(x_r))*100
                print('x1=',x1, 'x2=',    x2,'x_r=', x_r,   abs(error),'\n')
    return
x=sp.Symbol('x')
func=sp.exp(-x)-x
bisection_method(0,1,func,0.01)
       

```

```
import sympy as sp
def regular_falsi_method(xinitial_1,xinitial_2,func,error_tolerance):
    x1=xinitial_1
    x2=xinitial_2
    func_x1=func.evalf(subs={x:x1})
    func_x2=func.evalf(subs={x:x2})
    x_r=x2-func_x2*((x1-x2)/(func_x1-func_x2))
    func_xr=func.evalf(subs={x:x_r})
    error=1000
    while abs(error)>=error_tolerance:
        if (func_xr)*(func_x1)<0:
            old_root=x_r
            x1=x1
            x2=x_r
            func_x1=func.evalf(subs={x:x1})
            func_x2=func.evalf(subs={x:x2})
            x_r=x2-func_x2*((x1-x2)/(func_x1-func_x2))
            func_xr=func.evalf(subs={x:x_r})
            error=((x_r-old_root)/(x_r))*100
            print('x1=',x1,'x2=',x2,'x_r=',x_r,'error=',abs(error),'\n')
        if abs(error)>=error_tolerance:
            if (func_xr)*(func_x2)<0:
                old_root=x_r
                x1=x_r
                x2=x2
                func_x1=func.evalf(subs={x:x1})
                func_x2=func.evalf(subs={x:x2})
                x_r=x2-func_x2*((x1-x2)/(func_x1-func_x2))
                func_xr=func.evalf(subs={x:x_r})
                error=((x_r-old_root)/(x_r))*100
                print('x1=',x1,'x2=',x2,'x_r=',x_r,'error',abs(error),'\n')
    return

x=sp.Symbol('x')
func=x*sp.exp(x)-1
regular_falsi_method(0,1,func,0.001)    
```

```
import sympy as sp
def successive_approximation(func,initial_x,error_tolerance):
    error=10000
    y=initial_x
    while abs(error)>error_tolerance:                                               # express the x in the form of function of x for this method
        old_root=y
        y=func.subs(x,old_root)
        error=old_root-y
        print('root',y,'error',error)
    return
x=sp.Symbol('x')
func=sp.exp(-x)
successive_approximation(func,0.5,0.001)
```
