```
def seperate_by_class(data,target):
    class_wise_data={}
    for i in range(len(data)):
        if target[i] not in class_wise_data:
            class_wise_data[target[i]]=[]
        class_wise_data[target[i]].append(data[i])
    return(class_wise_data)

```

```
def mean(column_wise_data):
    sum=0
    for i in range(len(column_wise_data)):
        sum=sum+column_wise_data[i]
    return(sum/(len(column_wise_data)))

                                                                                 #across a particular class
```

```
def std(column_wise_data):
    std=0
    avg=mean(column_wise_data)
    for i in range(len(column_wise_data)):
        std+=((column_wise_data[i]-avg)**2/float((len(column_wise_data)-1)))
    return(avg,std,len(column_wise_data))


```

```
import numpy as np

def summarize_by_attributes(class_wise_data): 
    summary_class_wise={}
    for i in class_wise_data:
        if i not in summary_class_wise:
            summary_class_wise[i]=[]
        for j in range(len(class_wise_data[0][0])):
            column_wise_data=np.array(class_wise_data[i])

            summary_class_wise[i].append(std(column_wise_data[:,j]))
    return summary_class_wise


                                                                                #across a particular class
```

```
from math import exp
from math import sqrt
from math import pi
def calculate_probabilities(x_column,mean,std):
    gaussian_probability=(exp(-(x_column-mean)**2/(2*std**2)))*sqrt(1/(2*pi))*(1/std)
    return gaussian_probability
def predict_class(test_example,summary):
    probabilities={}
    for i in summary:
        prob=(summary[i][1][2])/(sum(summary[i][1][2] for i in summary))
        for j,c in enumerate(summary[i]):
            prob=prob*calculate_probabilities(test_example[j],c[0],c[1])
        probabilities[i]=prob
    return probabilities
def predictions(test_data,summary):
    predictions=[]
    for i in range(len(test_data)):
        probs=predict_class(test_data[i],summary)
        predictions.append(max(probs,key=probs.get))
    return predictions

    
def accuracy(predictions,targets):
    correct=0
    for i in range(len(targets)):
        if targets[i]==predictions[i]:
            correct=correct+1
    return((correct/len(targets))*100)
         
    
        


```
